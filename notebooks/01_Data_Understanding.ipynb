## Missing Value Analysis

Missing values are one of the most common challenges in real-world datasets.

Understanding the percentage of missing values helps determine the appropriate preprocessing strategy.

```python
missing = pd.DataFrame({
    "Missing Values": df.isnull().sum(),
    "Percentage": (df.isnull().sum()/len(df))*100
})

missing = missing.sort_values("Percentage", ascending=False)

missing
```

```python
missing[missing["Missing Values"] > 0]
```

## Duplicate Record Analysis

Duplicate records can introduce bias into the analysis and machine learning models.

This section identifies duplicate observations.

```python
duplicates = df.duplicated().sum()

print(f"Duplicate Rows : {duplicates}")
```
