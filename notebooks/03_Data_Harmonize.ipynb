## Phase 3 : Data Harmonizing

### Objective

The objective of this notebook is to standardize categorical variables across all OSMI survey years after data cleaning.

Although duplicate records, missing values, and invalid observations have already been addressed, survey responses collected across different years still contain inconsistencies such as:

- Different spellings
- Different capitalization
- Abbreviations
- Multiple representations of the same response
- Survey-specific response options

The harmonization process ensures that equivalent responses share a common representation while preserving the original meaning of the data.

The resulting dataset will be used for Exploratory Data Analysis (EDA), Feature Engineering, and Machine Learning.

```python
import pandas as pd
import numpy as np

pd.set_option("display.max_columns", None)

df = pd.read_csv("../data/processed/cleaned_osmi_data.csv")

print(df.shape)

df.head()
```

## Inspect Survey Years

Since this dataset combines multiple OSMI surveys, we first identify the survey years present in the data.

This helps us understand which records belong to each survey and allows year-specific harmonization where necessary.

Different survey years often use different response formats for the same question.

Inspecting each categorical variable separately for every survey year helps identify inconsistencies before harmonization.

## General Harmonization Rules

Some inconsistencies appear across multiple survey years.

These are standardized using common mapping dictionaries.

```python
binary_map = {

    "yes":"Yes",
    "Yes":"Yes",
    "YES":"Yes",

    "true":"Yes",
    "True":"Yes",

    "1":"Yes",
    "1.0":"Yes",

    "no":"No",
    "No":"No",
    "NO":"No",

    "false":"No",
    "False":"No",

    "0":"No",
    "0.0":"No"
}
```

```python
binary_map = {

    "Yes": "Yes",
    "yes": "Yes",
    "YES": "Yes",

    "True": "Yes",
    "true": "Yes",
    True: "Yes",

    "1": "Yes",
    "1.0": "Yes",
    1: "Yes",
    1.0: "Yes",

    "No": "No",
    "no": "No",
    "NO": "No",

    "False": "No",
    "false": "No",
    False: "No",

    "0": "No",
    "0.0": "No",
    0: "No",
    0.0: "No"
}

binary_columns = [

    "self_employed",
    "tech_company",
    "primary_role_tech",
    "sought_treatment",

    "discussed_with_employer",
    "discussed_with_coworkers",
    "coworker_discussed_with_you",

    "coverage_mental_health",

    "prev_tech_company",
    "prev_discussed_with_employer",
    "prev_discussed_with_coworkers",

    "openly_identified",
    "identified_affected_career"
]

for col in binary_columns:

    if col in df.columns:

        df[col] = (
            df[col]
            .replace(binary_map)
        )
```

## Standardize "Don't Know" Responses

Different survey years recorded uncertain responses using slightly different wording.

These responses are standardized to a single category.

```python
dontknow_map = {

    "Don't know": "Don't Know",
    "Don't Know": "Don't Know",
    "I don't know": "Don't Know",
    "Dont know": "Don't Know",
    "Not sure": "Don't Know",
    "Unsure": "Don't Know"
}

dontknow_columns = [

    "benefits",
    "know_options",
    "formal_discussion",
    "resources",
    "anonymity",
    "leave_ease",

    "family_history",

    "currently_disordered",
    "past_disorder",

    "prev_benefits",
    "prev_formal_discussion",
    "prev_resources",
    "prev_anonymity",
    "prev_comfort_supervisor"
]

for col in dontknow_columns:

    if col in df.columns:

        df[col] = df[col].replace(dontknow_map)
```

## Harmonize Gender

The gender variable contains a large number of synonymous responses, spelling variations, abbreviations, and free-text entries.

To improve interpretability while preserving meaningful distinctions, responses are grouped into four categories:

- Male
- Female
- Other
- Unknown

```python
# Gender Mapping
gender_map = {

    # Male
    "male": "Male",
    "m": "Male",
    "man": "Male",
    "cis male": "Male",
    "cis-male": "Male",
    "cisgender male": "Male",
    "male (cis)": "Male",
    "male-ish": "Male",
    "mail": "Male",
    "malel": "Male",
    "masculine": "Male",
    "identify as male": "Male",
    "cishet male": "Male",
    "make": "Male",
    "dude": "Male",
    "i have a penis": "Male",
    "man?": "Male", 
    "male (cisgender)": "Male",  
    "male (trans)": "Male",  
    
    # Female
    "female": "Female",
    "f": "Female",
    "woman": "Female",
    "cis female": "Female",
    "cis-female": "Female",
    "cisgender female": "Female",
    "female (cis)": "Female",
    "female (cisgender)": "Female",
    "female-identified": "Female",
    "femail": "Female",
    "femalw": "Female",
    "femile": "Female",
    "woman?": "Female",  
}

# Apply the mapping with a fallback to "Other"
def clean_gender(val):
    if pd.isna(val):
        return "Unknown"
    
    # Convert to string, strip, lowercase
    val = str(val).strip().lower()
    
    # Return mapped value, or "Other" if not found
    return gender_map.get(val, "Other")

# Apply to the dataframe
df["gender"] = df["gender"].apply(clean_gender)

# Check the results
print(df["gender"].value_counts())
```

## Harmonize Country Names

Country names are standardized to eliminate duplicate categories caused by abbreviations and alternate naming conventions.

```python
country_map = {

    "united states of america":"United States",
    "united states":"United States",
    "usa":"United States",
    "us":"United States",

    "uk":"United Kingdom",
    "england":"United Kingdom",
    "great britain":"United Kingdom",

    "brasil":"Brazil",

    "deutschland":"Germany"
}

country_columns = [

    "country_live",
    "country_work"
]

for col in country_columns:

    if col in df.columns:

        df[col] = (

            df[col]
              .astype("string")
              .str.strip()
              .str.lower()
              .replace(country_map)
              .str.title()

        )
```

## Year-Specific Harmonization

Some response options were introduced or modified in specific survey years.

These survey-specific values are harmonized individually while preserving their original meaning.

```python
# 2017 onwards

mask = df["source_file"] >= 2017

df.loc[mask, "benefits"] = (

    df.loc[mask, "benefits"]

      .replace({

          "I don't know":"Don't Know",
          "Not eligible for coverage / NA":"Not Eligible"

      })

)

df.loc[mask, "leave_ease"] = (

    df.loc[mask, "leave_ease"]

      .replace({

          "Difficult":"Very Difficult"

      })

)
```
